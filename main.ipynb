## Annual Breakdown of Shortages

Now, we have to analyze shortages by year, since we have data from many years of transactions.

First, we have to parse the **Payment Due Date** column because its dtype is *object*, which probably means the values are strings. Parsing it will help us to extract the year from each date.

```python
df["Parsed Due Date"] = pd.to_datetime(df["Payment Due Date"], format='%d/%m/%Y')

df["Payment Due Year"] = df["Parsed Due Date"].dt.year # Extract year from each date and put it on a new column

result = df.groupby("Payment Due Year")["Quantity Variance Amount"].sum() # Sum every shortage and group the results by year

print(result)

print(f"Total shortage: {result.sum()}") # Just to make sure we get the same result as above
```

Perfect. We got the result we were expecting (the sum of every yearly value is equal to the total that we previously calculated), but the numbers are worrying: they are growing over time. Let's see this graphically:

```python
import matplotlib.pyplot as plt

result.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Quantity Variance Amount (USD)')
plt.title('Annual Breakdown of Shortages')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
cumulative = result.cumsum()

plt.figure(figsize=(10, 6))
plt.plot(cumulative.index, cumulative.values, marker='o', linewidth=2, markersize=8, color='red')
plt.fill_between(cumulative.index, cumulative.values, alpha=0.3, color='red')
plt.xlabel('Year')
plt.ylabel('Cumulative Shortage Amount (USD)')
plt.title('Cumulative Shortage Growth')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

This upgoing trend is alarming. Let's analyze the distribution of the shortage amounts to identify patterns, outliers and anything else.

### Distribution Analysis of Shortage Amounts

```python
shortage_stats = df["Quantity Variance Amount"].describe(percentiles=[.25, .5, .75, .90, .95, .99])
print("Detailed Statistics of Shortage Amounts:")
print(shortage_stats)
print(f"\nMedian: ${shortage_stats['50%']:.2f}")
print(f"95th percentile: ${shortage_stats['95%']:.2f}")
print(f"99th percentile: ${shortage_stats['99%']:.2f}")
```

Just by seeing this, we may assume that there are **big** outliers. The biggest shortage recorded is almost 60k USD and the 99th percentile doesn't even reach a hundred dollars!

Let's take a look.

```python
above_99 = df[df["Quantity Variance Amount"] > shortage_stats['99%']]

above_99["Quantity Variance Amount"].describe()
```

We see that we have 436 values greater than percentile 99, but the same thing happens: there's a big jump between percentile 75 and the max value. Let's consider that this percentile 75 is our "outlier threshold":

```python
outliers = above_99[above_99["Quantity Variance Amount"] > 609.122500]

outliers_shortage = outliers["Quantity Variance Amount"]

plt.figure(figsize=(12, 6))
plt.boxplot(outliers_shortage, vert=False, widths=0.5)
plt.xlabel('Shortage Amount (USD)')
plt.title(f'Box Plot of Extreme Outliers (n={len(outliers_shortage)})')
plt.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()

print(f"Count: {len(outliers_shortage)}")
print(f"Total amount: ${outliers_shortage.sum():.2f}")
print(f"Percentage of total shortage: {outliers_shortage.sum()/total_shortage_amount*100:.2f}%")
print(f"Mean outlier value: ${outliers_shortage.mean():.2f}")
print(f"Median outlier value: ${outliers_shortage.median():.2f}")
print(f"Largest outlier: ${outliers_shortage.max():.2f}")
```

This is **huge**. Just 109 out of 43546 entries constitute almost 80% of the total shortage! That's a 0.25% of our dataset.

What about the dates? From what we saw earlier, we should expect those values are from 2024 and 2025.

```python
outliers_count = outliers["Payment Due Year"].value_counts().sort_index()

outliers_count.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Amount of Outliers')
plt.title('Outliers by Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

for year in outliers_count.index:
    year_with_shortage = df[(df["Payment Due Year"] == year) & (df["Quantity Variance Amount"] > 0)]
    year_shortage_count = len(year_with_shortage)
    year_outliers = outliers_count[year]
    percentage = (year_outliers / year_shortage_count) * 100
    print(f"{year}: {year_outliers} outliers out of {year_shortage_count} transactions with shortage ({percentage:.2f}%)")
```

Now, we can confirm that the problem relies almost entirely on the outliers. Only 5% of the transactions with sortage made in 2024 are our outliers, even though they represent the vast majority of the money! I think all the future work should focus on this!
